```python

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("dark")
ls=Match_df['venue'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10,4))

temp =sns.barplot(ls.index, ls.values, alpha=0.8)
plt.title('MOST PLAYED VENUE')
plt.ylabel('COUNT', fontsize=14)
plt.xlabel('NAME OF THE STADIUMS', fontsize=15)
temp.set_xticklabels(rotation=90,labels=ls.index,fontsize=8)
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("dark")
ls=Match_df['winner'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(10,4))

temp =sns.barplot(ls.index, ls.values, alpha=0.8)
plt.title('Points Table')
plt.ylabel('Points', fontsize=14)
plt.xlabel('Teams', fontsize=15)
temp.set_xticklabels(rotation=90,labels=ls.index,fontsize=10)
plt.show()
```

```python
temp=pd.concat([Match_df['umpire1'],Match_df['umpire2'],Match_df['umpire3']]).value_counts().sort_values(ascending=False)
plt.figure(figsize=(20,5))
Most_umpired =sns.barplot(x=temp.index, y=temp.values, alpha=0.9)

plt.title('Favorite umpire')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Name of the Umpire', fontsize=15)
Most_umpired.set_xticklabels(rotation=90,labels=temp.index,fontsize=15)
plt.show()
```

```python
Dubai_stadium=Match_df.loc[(Match_df['venue']=='Dubai International Cricket Stadium') ]
Dubai_stadium_win_by_runs=Dubai_stadium[Dubai_stadium['win_by_runs']>0]
slices=[len(Dubai_stadium_win_by_runs),len(Dubai_stadium)-len(Dubai_stadium_win_by_runs)]
labels=['Batting first','Batting Second']
plt.pie(slices,labels=labels,startangle=20,shadow=1,explode=(0,0.2),autopct='%2.3f%%',colors=['red','green'])
plt.title('Dubai International Cricket Stadium')
plt.show()
```
